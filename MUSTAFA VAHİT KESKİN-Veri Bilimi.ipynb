#difference():iki kümenin farkı ya da (-)
#intersection():iki kümenin kesişimi ya da(&)
#union():ikş kümenin birleşimi
symmetric difference():ikisinde de olmayanaları

```python
set1=set([1,2,3,5])
set2=set([22,34,1,2,3])
```

```python
set1.difference(set2)
```

```python
set2.difference(set1)
```

```python
set1.symmetric_difference(set2)
```

```python
set2.symmetric_difference(set1)
```

```python
set1-set2
```

```python
set2-set1
```

```python
set1.intersection(set2)
set2.intersection(set1)
kesisim=set1&set2
kesisim
```

```python
set1.intersection_update(set2)
set1
```

```python
set2.intersection_update(set1)
set2
```

```python
set3=set([7,8,9])
set4=set([5,6,7,8,9,10])
set3.isdisjoint(set4)
#iki kümenin kesişimi boş olup olmadığı 
#isdisjoint metodu ile sorgulanır
```

```python
set3.issubset(set4)
#altküme
```

```python
set4.issuperset(set3)
#kapsayıclık
```
